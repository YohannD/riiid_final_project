```python
import pandas as pd
```

```python
train=pd.read_csv('../data/train.csv')
```

# Splitting on users

```python
long=len(train)
```

```python
cut=int(long/100)
```

```python
cuts=[0]
for i in range(99):
    cuts.append(train.iloc[i*cut:(i+1)*cut].query('timestamp==0').iloc[-1].row_id)
cuts.append(train.iloc[-1].row_id+1)
```

```python
for i in range(100):
    print(f'cut {i}')
    subtrain=train.iloc[cuts[i]:cuts[i+1]]
    subtrain.to_csv(f'../data/cuts/train_user_cut_{i}')
```
